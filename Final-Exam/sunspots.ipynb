```python
import numpy as np

#file=open('ISSN_D_tot.csv','r')
#dt=np.dtype()
data=np.recfromcsv('ISSN_D_tot.csv',delimiter=',',dtype=float)

data=np.array(data)
#latest=data[0][0]
#date=data[0]
#for x in range (0,len(data)):
#    if latest<data[x][3]:
#        latest=data[x][3]
#        date=data[x]
print("The most recent sunspot observation were on " + str(int(data[-1][0])) + "-"+str(int(data[-1][1]))+"-"+str(int(data[-1][2])))
```

```python
from scipy import stats
print("Statistics of daily sunspot observations (1818-present):")
observations=[]
for x in range (0,len(data)):
    if data[x][4] != -1:
        observations.append(data[x][4])

observations=np.array(observations)
mean=np.mean(observations)
median=np.median(observations)
deviation=np.std(observations)
print("Mean: " + str(mean))
print("Median: "+str(int(median)))
print("Standard deviation: "+str(deviation))
```

```python
import matplotlib.pyplot as plt

times=[]
for x in range (0,len(data)):
    if data[x][4] != -1:
        times.append(data[x][3])

times=np.array(times)
plt.title('Daily sunspot observations (1818 to 2014)')
plt.xlabel('Time')
plt.ylabel('Sunspot observation')
plt.plot(times,observations,'co')

x=np.linspace(1800,2050)
plt.plot(x,mean,color='blue')
plt.plot(x,median,color='black')
plt.plot(x,mean+deviation,color='magenta',dash=True)
plt.plot(x,mean-deviation,color='magenta',dash=True)

savefig('sunplot_plot.png',dpi=48)
plt.show()
```
